```python
np.random.seed(1234)

# Take a random sample of 25 observations
sample = raw_data.sample(n=25)
```

```python
# Calculate mean and highest Glucose values for the sample
sample_mean_glucose = sample['Glucose'].mean()
sample_highest_glucose = sample['Glucose'].max()

# Calculate mean and highest Glucose values for the population
population_mean_glucose = raw_data['Glucose'].mean()
population_highest_glucose = raw_data['Glucose'].max()


print("Sample Mean Glucose:", sample_mean_glucose)
print("Sample Highest Glucose:", sample_highest_glucose)
print("Population Mean Glucose:", population_mean_glucose)
print("Population Highest Glucose:", population_highest_glucose)
```

```python
# Create a bar chart for comparison
labels = ['Sample Mean Glucose', 'Sample Highest Glucose', 'Population Mean Glucose', 'Population Highest Glucose']
values = [sample_mean_glucose, sample_highest_glucose, population_mean_glucose, population_highest_glucose]

plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['blue', 'orange', 'green', 'red'])
plt.ylabel('Glucose Values')
plt.title('Comparison of Glucose Statistics: Sample vs Population')
plt.show()
```

```python
#c) 


def create_bootstrap_samples(raw_data, n_samples=500, sample_size=150):
    bootstrap_samples = []
    for _ in range(n_samples):
        sample = raw_data.sample(n=sample_size, replace=True)
        bootstrap_samples.append(sample)
    return bootstrap_samples

```

```python
bootstrap_samples_bp = create_bootstrap_samples(raw_data['BloodPressure'], n_samples=500, sample_size=150)

# Calculate statistics for each bootstrap sample
bootstrap_means_bp = [sample.mean() for sample in bootstrap_samples_bp]
bootstrap_stds_bp = [sample.std() for sample in bootstrap_samples_bp]
bootstrap_percentiles_bp = [np.percentile(sample, 95) for sample in bootstrap_samples_bp]

# Calculate mean, standard deviation, and percentile for BloodPressure in the population
population_mean_bp = raw_data['BloodPressure'].mean()
population_std_bp = raw_data['BloodPressure'].std()
population_percentile_bp = np.percentile(raw_data['BloodPressure'], 95)

# Print the statistics
print("Average Mean of Bootstrap Samples for BloodPressure:", np.mean(bootstrap_means_bp))
print("Average Standard Deviation of Bootstrap Samples for BloodPressure:", np.mean(bootstrap_stds_bp))
print("Average 95th Percentile of Bootstrap Samples for BloodPressure:", np.mean(bootstrap_percentiles_bp))
print("Population Mean BloodPressure:", population_mean_bp)
print("Population Standard Deviation BloodPressure:", population_std_bp)
print("Population 95th Percentile BloodPressure:", population_percentile_bp)
```

```python
# Create comparison charts for mean, std, and percentile of BloodPressure
labels = ['Population Mean BP', 'Bootstrap Mean BP', 'Population Std BP', 'Bootstrap Std BP',
          'Population 95th Percentile BP', 'Bootstrap 95th Percentile BP']
values = [population_mean_bp, np.mean(bootstrap_means_bp), population_std_bp, np.mean(bootstrap_stds_bp),
          population_percentile_bp, np.mean(bootstrap_percentiles_bp)]

plt.figure(figsize=(10, 6))
plt.barh(labels, values, color=['blue', 'orange', 'green', 'red', 'purple', 'pink'])
plt.xlabel('BloodPressure Values')
plt.title('Comparison of BloodPressure Statistics: Bootstrap vs Population')
plt.show()
```
